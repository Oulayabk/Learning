# linear regression

```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

## 1. Dataset

```python
x,y = make_regression(n_samples=100, n_features=1, noise =10)
plt.scatter(x,y)
```

```python
print(x.shape)
y = y.reshape(y.shape[0],1)
print(y.shape)
```

```python
# créer la matrice X
X = np.hstack((x,np.ones(x.shape)))
X.shape
```

```python
# créer Theta0

theta=np.random.randn(2,1)
theta
```

## 2. Model

```python
# créer la fonction model

def model(X,theta):
    return X.dot(theta)
```

```python
plt.plot(x, model(X,theta))
```

```python
# test
model(X,theta)
plt.scatter(x,y)
plt.plot(x, model(X,theta))
```

## 3. cost function

```python
def cost_function (X, y, theta): 
    m = len(y)
    return 1/(2*m) * np.sum((model(X, theta) - y)**2)
```

```python
cost_function(X, y, theta)
```

## 4. gradients and Gradient descent

```python
def grad(X, y, theta):
    m = len(y)
    return 1/m * X.T.dot(model(X, theta) - y)
```

```python
def grad_descent(X, y, theta, learning_rate, n_iterations):
    for i in range(n_iterations):
        theta = theta - learning_rate*grad(X, y, theta)
    return theta
```

## 5. model training

```python
theta_final = grad_descent(X, y, theta, learning_rate=0.01, n_iterations=1000)
```

```python
theta_final
```

```python
cost_function(X,y,theta_final)
```

```python
predictions = model(X, theta_final)
```

```python
# test predictions 1
plt.scatter(x,y)
plt.plot(x, predictions, c='r')
```
